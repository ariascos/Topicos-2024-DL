## Clases en Python

En Python, una clase es una plantilla para la creación de objetos. Estos objetos son instancias de la clase, y cada instancia puede tener sus propios atributos (variables) y métodos (funciones). Las clases son una forma de encapsular datos y funcionalidad relacionada en un solo objeto.
Características de una clase:

*    Atributos: Son variables que pertenecen a la clase y al objeto. Cada objeto puede tener sus propios valores para estos atributos.

*    Métodos: Son funciones definidas dentro de la clase que pueden acceder y manipular los atributos de la clase y del objeto.

*    Constructor: Es un método especial llamado __init__ que se utiliza para inicializar los atributos de la clase cuando se crea un nuevo objeto.

*    Encapsulación: Las clases en Python pueden tener métodos y atributos públicos, protegidos o privados para controlar el acceso a los datos y la funcionalidad de la clase.

````python
class Persona:
    def __init__(self, nombre, edad, genero):
        self.nombre = nombre
        self.edad = edad
        self.genero = genero

    def presentarse(self):
        return f"Hola, soy {self.nombre}, tengo {self.edad} años y soy {self.genero}."

    def cumplir_anios(self):
        self.edad += 1
        return f"¡Feliz cumpleaños! Ahora tengo {self.edad} años."

# Crear una instancia de la clase Persona
persona1 = Persona("Juan", 30, "masculino")

# Acceder a los atributos y métodos de la persona
print(persona1.presentarse())  # Salida: Hola, soy Juan, tengo 30 años y soy masculino.
print(persona1.cumplir_anios())  # Salida: ¡Feliz cumpleaños! Ahora tengo 31 años.
````
